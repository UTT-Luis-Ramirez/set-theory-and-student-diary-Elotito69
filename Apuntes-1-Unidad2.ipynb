**Coleciones**:

 Tienen elementos. son una agrupacion de cosas con elementos en comun.

**Informacion extra:**

Es simplemente un grupo de eventos o situaciones que se agrupan o se consideran juntas.

Por ejemplo, imagina que tienes una bolsa con diferentes bolas de colores. Cada bola es un evento (por ejemplo, sacar una bola roja, sacar una bola azul, etc.). Si consideras todas las posibles bolas que puedes sacar de la bolsa, eso sería una colección de eventos.

**Elementos:**

 Caso abstracto y lo define la persona que lo esta analizando y es cualquier cosa.

**Informacion extra**:

Es un resultado posible de un experimento o situación aleatoria.

Por ejemplo, si tiras un dado, los elementos posibles son los números que pueden salir: 1, 2, 3, 4, 5 o 6. Cada uno de esos números es un elemento en el espacio muestral, que es el conjunto de todos los posibles resultados.
